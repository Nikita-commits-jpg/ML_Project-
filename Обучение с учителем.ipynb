## Создание собственных функций для предобработки данных.

Прежде чем выполнять предобработку в данных, создали функции для облегчения работы над проектом: 

```python
#Функция для трансформации значений в колонках в необходимый формат
def columns_to_transform(DF):
    DF.columns = DF.columns.str.lower().str.replace(' ', '_')
```

```python
#Функция для автоматического вывода на экран следующих методов: info(), sample(), describe(), shape, dtypes, isna()
def overall_info(DF):
    
    # Выводим информацию о датафрейме
    print('Вывод общей информации о данных')
    display(DF.info())
    print()

    # Выводим случайную выборку строк
    print('Вывод пяти случайных строк DataFrame')
    display(DF.sample(5))
    print()

    # Выводим описание числовых столбцов
    print('Вывод числового описания данных')
    display(DF.describe())
    print()

    # Выводим форму датафрейма
    print('Вывод данных о размере DataFrame')
    display(DF.shape)
    print()

    # Выводим информацию о типах данных столбцов
    print('Вывод типов данных столбцов DataFrame')
    display(DF.dtypes)
    print()

    # Выводим информацию о пропущенных значениях
    print('Вывод информации о наличии пропусков в DataFrame')
    display(DF.isna().sum())
```

```python
#Функция для автоматического вывода на экран явных дубликатов
def ambigious_duplicates(DF):
    display(DF.duplicated().sum())
```

```python
#Функция для автоматического вывода на экран метода value_counts() с целью анализа наличия неявных дубликатов
def unambigious_duplicates(DF):
    for column in DF.columns:
        if DF[column].dtype == 'object': 
            display(DF[column].value_counts())
```

```python
#Функция для автоматического устранения неявных дубликатов
def drop_unambigious_duplicates(DF, column, to_be_replaced, new_value):
    DF[column] = DF[column].apply(lambda x: x.replace(to_be_replaced, new_value))
```

```python
#Функция для автоматической идентификацией столбцов с выбросами
def analyze_anomalies(DF):
    for column in DF.columns:
        if DF[column].dtype in [np.int64, np.float64]:
            Q1 = DF[column].quantile(0.25)
            Q3 = DF[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            anomalies = DF[(DF[column] < lower_bound) | (DF[column] > upper_bound)]
            if not anomalies.empty:
                print(f"Аномальные значения в столбце {column}:")
                display(DF[column].describe())
                # Дополнительное построение графиков для визуализации выборосов и аномалий
                plt.figure(figsize=(10, 6))
                sns.boxplot(x=DF[column])
                plt.title(f"Диаграмма размаха для столбца {column}")
                plt.xlabel(column)
                plt.show()
```

```python
#Функция для создания объединенных графиков вида pie_chart:
def plot_pie_charts(df, features):
    if isinstance(features, str):
        features = [features]

    num_charts = len(features)
    fig, axes = plt.subplots(1, num_charts, figsize=(num_charts * 8, 6))

    for i, feature in enumerate(features):
        counts = df[feature].value_counts()
        labels = counts.index.tolist()
        sizes = counts.values
        colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
        axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
        axes[i].axis('equal')
        axes[i].set_title(f'Круговая диаграмма признака {feature}')

    plt.show()
```

```python
#Функция для создания объединенных графиков вида bar_chart:
def plot_bar_chart(data, columns, figsize=(15, 5), style='default'):
    plt.style.use(style)

    num_plots = len(columns)
    cols = num_plots  
    rows = 1  

    fig, axes = plt.subplots(rows, cols, figsize=figsize)

    for i, column in enumerate(columns):
        counts = data[column].value_counts()

        ax = axes[i]
        counts.plot(kind='bar', ax=ax)
        ax.set_title(f'Столбчатая диаграмма для {column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Количество')

    plt.tight_layout()  
    plt.show()
```

```python
#Функция для создания объединенных графиков типа boxplot_chart:
def plot_hist_and_boxplot(df, features):
    if isinstance(features, str):
        features = [features]

    num_charts = len(features)
    fig, axes = plt.subplots(2, num_charts, figsize=(num_charts * 6, 12))

    for i, feature in enumerate(features):
        # Гистограмма
        sns.histplot(data=df, x=feature, kde=True, ax=axes[0, i])
        axes[0, i].set_title(f'Гистограмма признака {feature}')
        axes[0, i].set_ylabel('Частота')

        # Ящик с усами
        sns.boxplot(data=df, y=feature, ax=axes[1, i])
        axes[1, i].set_title(f'Ящик с усами признака {feature}')

    plt.tight_layout()
    plt.show()
```

```python
#Функция для создания одного графика типа pie_chart:
def one_pie_chart(table, column, title):
    values = table[column].value_counts()
    values.plot(
        radius=1.0,
        figsize=(15, 6),
        kind='pie',
        startangle=90,  
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],
        labels=values.index,
        title=title,
        autopct='%1.1f%%',
        rot=0  
    )
    return 'На иллюстрации ниже изображен график типа круговая диаграмма'
```

```python
#Функция для создания одного графика типа bar_chart:
def one_bar_chart(table, column, title):
    
    values = table[column].value_counts()
    values.plot(
            figsize=(15, 6),
            kind='bar',
            title=title)
    plt.ylabel('Частота')
 
    return 'На иллюстрации ниже изображен график типа круговая диаграмма' 
```

```python
def one_hist_chart(table, column, title):
    sns.histplot(data=table, x=column, kde=True)
    plt.ylabel('Частота')
    plt.title(title) 
    plt.show()
```

```python
def one_boxplot_chart(table, column, title, num1, num2):
    plt.figure(figsize=(num1, num2))
    sns.boxplot(data=table, x=column)
    plt.title(title)
    plt.xlabel('Распределение значений')
    plt.grid()
    plt.xticks(rotation=0)
    plt.show()
```

## Анализ признаков из общего дата-сета на мультиколлиниарность.

Для анализа мультиколлиниарности признаков воспользовались мерой VIF:

```python
#Добавление константы в DataFrame
combined_data_with_const = add_constant(combined_data)
```

```python
#Фильтрацию только по числовым значениям
num_columns = combined_data_with_const.select_dtypes(include='number').columns
```

```python
#Создание функции подсчета VIF - calc_vif
def calc_vif(df):

    vif_data = pd.DataFrame()
    vif_data["variables"] = df.columns
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

    return vif_data
```

```python
#Рассчет VIF при помощи функции calc_vif
calc_vif(combined_data_with_const[num_columns].drop('id', axis=1)).sort_values(by='VIF', ascending=False)
```

Ключевые пороги для VIF:

- VIF < 1: минимальная мультиколлинеарность;


- 1 ≤ VIF ≤ 5: признак умеренной мультиколлинеарности;


- VIF > 5 (или 10): Сильная мультиколлинеарность, вероятно, данные коррелируют с другими;

Согласно коэффициенту VIF - мультколлинеарность между признаками отсутствует.

**Промежуточные выводы по анализу признаков из общего дата-сета на мультиколлиниарность:**

# Использование пайплайнов.

## Подготовка данных к машинному обучению (создание пайплайнов).

Согласно корреляционному анализу мультиколлиниарность отсутствует, однако присутствуют устойчивые линейные связи между **выручка_предыдущий месяц и выручка_текущий_месяц**. В виду этого, приняли решение удалить один из признаков для получения более высоких значений метрик.

```python
#Создание переменных с константами
RANDOM_STATE = 42
TEST_SIZE = 0.25
```

```python
#Разделение выборок 
X = combined_data.drop(['покупательская_активность', 'id', 'выручка_предыдущий_месяц'], axis=1)
y = combined_data['покупательская_активность']
```

В виду того, что распредеелние целевого признака неравномерное - добавили в функцию train_test_split дополнительный параметр - stratify.

```python
#Разделение данных на тренировочную и тестовую выборки 
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y, 
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE, 
    stratify = y
)
```

```python
# Создаем экземпляр LabelEncoder
le = LabelEncoder()

# Приводим целевой признак y_train к кодированному виду и сохраняем результат в y_train_encoded
y_train_encoded = le.fit_transform(y_train)

# Кодируем целевой признак y_test с использованием уже обученного LabelEncoder и сохраняем результат в y_test_encoded
y_test_encoded = le.transform(y_test)
```

```python
#Создание списков с названиями признаков
ohe_columns = ['популярная_категория']              


ord_columns = [
    'тип_сервиса', 'разрешить_сообщать']
```

```python
num_columns = ['маркет_актив_6_мес', 
               'маркет_актив_тек_мес', 'длительность',
              'акционные_покупки', 'средний_просмотр_категорий_за_визит', 
               'неоплаченные_продукты_штук_квартал',
               'ошибка_сервиса',
               'страниц_за_визит', 'выручка_препредыдущий_месяц', 
               'выручка_текущий_месяц', 'время_предыдущий_месяц', 
               'время_текущий_месяц']
```

```python
#Создание пайплайнов для категориальных переменных

#1
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
)
```

```python
#2
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум'],
                    ['нет', 'да']
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
#Создание общего пайплайна для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough')

#Создание итогового пайплайна
pipe_final= Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE, 
                                          max_depth=5, 
                                          min_samples_split=25, 
                                          min_samples_leaf=25, 
                                          max_features='log2'))
    ]
)
```

```python
pipe_final.fit(X_train, y_train_encoded)
```

## Настройка и подготовка моделей.

```python
#Использование переменной param_grid для дальнейшего использования в GridSearchCV
param_grid = [
    
    # словарь для модели DecisionTreeClassifier()
    {
    'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
    'models__max_depth': [3, 5],  # Сократили количество значений (2 коммит)
    'models__min_samples_split': [2, 5],  # Сократили количество значений (2 коммит)
    'models__min_samples_leaf': [1, 3],  # Сократили количество значений (2 коммит)
    'preprocessor__num': [StandardScaler(), 'passthrough']  # Оставили два наиболее часто используемых масштабировщика (2 коммит)
    },
    
    # словарь для модели KNeighborsClassifier() 
   {
    'models': [KNeighborsClassifier()],
    'models__n_neighbors': range(1, 10),
    'models__weights': ['uniform', 'distance'],
    'models__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'models__leaf_size': [10, 30, 50],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='newton-cg', max_iter=2000)],
        'models__penalty': ['l2'],
        'models__C': [1, 10, 100],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'] 
    },
    
    # словарь для модели SVC()
    {
    'models': [SVC(random_state=RANDOM_STATE, probability=True)],
    'models__C': [1, 10, 100],  # Сократили количество значений (1 коммит)
    'models__kernel': ['linear', 'rbf'], 
    'models__gamma': ['scale', 'auto', 0.1],  # Сократили количество значений (1 коммит)
    'models__degree': [2, 3],
    'preprocessor__num': [StandardScaler(), 'passthrough']  # Оставили два наиболее часто используемых масштабировщика (1 коммит)
    }
]
```

Касательно выбора метрики, **главной** метрикой в поставленной задачи является метрика **recall**, так как необходимо предсказать вероятность того, что покупатель перейдет в категорию **снизилась**. Остальные метрики также были посчитаны - для дополнительной информации.

```python
#Использование метода GridSearchCV для поисков лучших гиперпараметров модели
random_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    scoring=['recall', 'precision', 'roc_auc', 'accuracy'],
    random_state = 5,
    refit='recall',
    n_jobs=-1
)
```

```python
#Обучение  random_search
random_search.fit(X_train, y_train_encoded)
```

```python
#Вывод на экран лучшей модели МО
display('Лучшая модель и её параметры:', random_search.best_estimator_)
```

```python
#Вывод результатов лучшей метрики при кросс-валидации
display(f'Лучшая метрика (recall) на кросс-валидации:, {round(random_search.best_score_ * 100, 2)}')
```

    - Порог, при котором вероятность отнесения данных к классу Снизилась считается высоковероятной более 0.8 - фильтрацию по результатам **моделирования**;
    - Второй критерий при **выделенном** сегменте является критерий прибыльности: здесь установили критерий более 4.0ед (выше медианной).

- Следующий шаг стал **исследовательский анализ данных**, по его результату получили следующие выводы:
    - В среднем основной объем пользователей зарегестрирован в промежутке от 400 до 600 дней;
    - Стоит отметить, что послее 800 дней происходит резкое снижение по количеству пользователей; 
    - Что касается количества разных взаимодействий по маркетингу - в среднем выделяется 4 взаимодействия. 
    - В виду высокой линейной связи между двумя признаками, связанными с выручкой - перед обучением модели удалили признак выручки за предыдущий месяц, провели анализ по оставшимеся двум другим месяцам; 
    - За 1-й месяц квартала средняя выручка наблюдается в районе 4500р, когда как в текущем месяце (3ом) данный показатель лучше - 6000руб;
    - Исходя из вышесказанного можно сделать вывод, что выручка по текущему месяцу больше чем за препредыдущий;
    - Меньшую заинтересованность клиентов также можно оценивать по количеству проведенного времени на сайте: в препредыдущем месяце оно составляло 10-11 мин, тогда как в текущем от 10 мин;
    - Кроме этого, большая плотность пользователей задерживается на сайте - видно выбросы более частое явление в текущем месяце.
    - Ожидаемо, пользователи со стандартной формой подписки привелигируют над премиумом сегментом - 66,7 против 33,3%;
    - Большинство пользователей (75,3%) разрешают, чтобы им присылали различного рода рассылки и другие формы рекламы.
    - В среднем пользователи просматривают 2 категории (40,7%); 3 категории также довольно популярны - 27,2%;
    - В текущем месяце маркетинг слабее, в сравнении с 6 месяцами в среднем - всего 4 взаимодействия на большинство клиентов (53,7%);
    - Касательно самых популярных категорий товаров (от наибольшего к ТОП-3): товары для детей (32.7%), косметика и аксессуары и домашний текстиль (на долю этих категорий приходится по 23,5% и 18,5% соответствено);
    - В основном клиенты не оплачивают от 3 до 5 товаров; очевидно, с ростом числа  неоплаченых товаров, снижается их процент выраженный в клиентах.

- **Самые главные выдержки и рекомендации:**
